```python
from datasets import load_dataset
import pandas as pd

dataset1 = load_dataset('financial_phrasebank', 'sentences_75agree')
dataset2 = load_dataset('FinanceInc/auditor_sentiment')
```

```python
# 문장과 레이블 추출
sentences1 = dataset1['train']['sentence']
labels1 = dataset1['train']['label']

sentences2 = dataset2['train']['sentence']
labels2 = dataset2['train']['label']

sentences3 = dataset2['test']['sentence']
labels3 = dataset2['test']['label']

# 문장과 레이블을 결합
sentences = sentences1 + sentences2 + sentences3
labels = labels1 + labels2 + labels3

# 문장과 레이블을 DataFrame으로 변환
df = pd.DataFrame({'sentence': sentences, 'label': labels})

# 중복 제거 전의 데이터셋 크기
original_length = len(df)

# 중복된 문장을 제거
df = df.drop_duplicates(subset=['sentence'])

# 중복 제거 후의 데이터셋 크기
deduplicated_length = len(df)

# 중복된 데이터의 개수 계산
num_duplicates = original_length - deduplicated_length
print(f"중복된 데이터의 개수: {num_duplicates}")

# 중복 제거된 문장과 레이블을 다시 리스트로 변환
sentences = df['sentence'].tolist()
labels = df['label'].tolist()

print(df)
```
